```
from azure.cognitiveservices.vision.computervision import ComputerVisionClient
from azure.cognitiveservices.vision.computervision.models import OperationStatusCodes
from azure.cognitiveservices.vision.computervision.models import VisualFeatureTypes
from msrest.authentication import CognitiveServicesCredentials
from array import array
import os
import sys
import time
```

```
''' Autenticación '''

#Computer vision info
subscription_key = "[key 1]" 
endpoint = "[endpoint]"
```

```
compuitervision_client = ComputerVisionClient(endpoint, CognitiveServicesCredentials(subscription_key))
```

```
imagen = open("/dbfs/mnt/blob_images/6a01a73dbd1d98970d01b7c9541a63970b.jpg", "rb") #Ubicación del archivo
```

```
images = spark.read.format("image").load("/mnt/blob_images/6a01a73dbd1d98970d01b7c9541a63970b.jpg")
display(images)
```

```
respuesta = compuitervision_client.read_in_stream(imagen, raw=True)
```

```
localizacion_operacion = respuesta.headers["Operation-Location"]
```

```
localizacion_operacion = localizacion_operacion.split("/")[-1]

# Call the "GET" API and wait for the retrieval of the results
while True:
  read_result = compuitervision_client.get_read_result(localizacion_operacion)
  if read_result.status.lower() not in ['notstarted', 'running']:
    break
  print('...')
  time.sleep(10)
    
#Print results, line by line
if read_result.status == OperationStatusCodes.succeeded:
  for text_result in read_result.analyze_result.read_results:
    for line in text_result.lines:
      print(line.text)
```
