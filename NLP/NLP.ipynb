```python
from nltk.util import ngrams

first_head = data.text[0:6]
word_tokenized = nltk.word_tokenize(''.join(first_head))
list(ngrams(word_tokenized, 3))
```

```python
def top_k(word_tokens, n, z):
  n_gram_list = list(ngrams(word_tokens, n))
  n_gram = [' '.join(x) for x in n_gram_list]
  n_gram_counter = Counter(n_gram)
  most_common_gram = n_gram_counter.most_common(z)
  print(most_common_gram)
```

```python
reviews_text = ''.join(reviews)

words = nltk.word_tokenize(reviews_text)

top_k(words, 1, 10)
```

Libreria FuzzyWuzzy para las contracciones de palabras

```python
def get_top_n_words(corpus, n=1, k=1):
  vec = CountVectorizer(ngram_range=(k, k), stop_words='english').fit(corpus)
  bad_of_words = vec.transform(corpus)
  sum_words = bad_of_words.sum(axis=0)
  words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
  words_freq = sorted(words_freq, key= lambda x: x[1], reverse=True)
  return (words_freq[:n])
```

```python
common_words = get_top_n_words(bad_comment['text'], 50, 2)

for word, freq in common_words:
  print(word, freq)
```
